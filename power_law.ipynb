```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=12345)
```

```python
risk_levels = np.array([1, 2, 3, 4, 5])
expected_returns = np.array([0.10, 0.30, 0.50, 0.80, 1.20])
std_returns = np.array([0.05, 0.15, 0.30, 0.60, 1.00])

n_obs_per_level = np.array([400, 250, 120, 50, 20])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

for i in range(len(risk_levels)):
    n_i = n_obs_per_level[i]
    actual_returns = rng.normal(loc=expected_returns[i],
                                 scale=std_returns[i],
                                 size=n_i)
    x_jitter = risk_levels[i] + rng.uniform(-0.15, 0.15, size=n_i)
    ax.scatter(x_jitter, actual_returns, alpha=0.3, color="grey", s=15)

ax.plot(risk_levels, expected_returns, color="blue",
        linewidth=2, marker="o", label="Expected return")

ax.set_xlabel("Risk level")
ax.set_ylabel("Return")
ax.set_title("Risk and dispersion of realized returns")
ax.axhline(0, color="black", linewidth=0.5)
ax.legend()
ax.grid(alpha=0.3)

plt.show()
```
